##### Step 1: Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


```

##### Step 2: Create grid of x and y values

```python

x = np.linspace(-3, 3, 60)
y = np.linspace(-3, 3, 60)
X, Y = np.meshgrid(x, y)

# np.linspace(-3, 3, 60) → makes 60 evenly spaced numbers from -3 to 3.
# np.meshgrid(x, y) → creates a grid of all possible (x, y) coordinate pairs

```

##### Step 3: Define the function f(x, y) = 4x²y + 2y³x

```python

Z = 4 * X**2 * Y + 2 * Y**3 * X
```

##### Step 4: Compute partial derivatives

```python

df_dx = 8 * X * Y + 2 * Y**3      # ∂f/∂x = formula = nx^n-1
df_dy = 4 * X**2 + 6 * Y**2 * X   # ∂f/∂y = formula = nx^n-1

```

##### Step 5: Plot the 3D surface

```python

fig = plt.figure(figsize=(10, 7)) # width, height
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)

# # add_subplot(nrows, ncols, index) one big plot 
# 111 means 1 → number of rows of plots1 → number of columns of plots, 1 → this is the first (and only) plot
# plot_surface() draws a smooth colored surface:
# cmap='viridis' adds nice color shading.
# alpha=0.8 makes it slightly transparent

ax.set_title('3D Surface and Gradient: f(x, y) = 4x²y + 2y³x', fontsize=13)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('f(x, y)')
plt.show()
```
