##### 1. Identify the smallest of three numbers

```python
# Given numbers
a = 12
b = 7
c = 15

# Find the smallest number
if a <= b and a <= c:
    smallest = a
elif b <= a and b <= c:
    smallest = b
else:
    smallest = c

print(f"The smallest number among {a}, {b}, {c} is {smallest}")
```

##### 13. Check whether a number is a perfect square (basic logic)

```python
# Given number
num = 36

# Initialize a flag
is_perfect_square = False

# Check perfect square
for i in range(1, num + 1):
    if i * i == num:
        is_perfect_square = True
        break

if is_perfect_square:
    print(f"{num} is a perfect square")
else:
    print(f"{num} is not a perfect square")
```

##### 19. Calculate electricity bill using conditional slabs

```python
# Given units consumed
units = 350
bill = 0

# Calculate bill based on slabs
if units <= 100:
    bill = units * 5        # Rs. 5 per unit for first 100 units
elif units <= 200:
    bill = 100 * 5 + (units - 100) * 7  # Rs. 7 per unit for next 100 units
elif units <= 300:
    bill = 100 * 5 + 100 * 7 + (units - 200) * 10  # Rs. 10 per unit for next 100 units
else:
    bill = 100 * 5 + 100 * 7 + 100 * 10 + (units - 300) * 15  # Rs. 15 per unit above 300

print(f"Units Consumed: {units}, Electricity Bill: Rs. {bill}")
```
