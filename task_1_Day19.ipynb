##### Step 1: Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

```

##### Step 2: Create grid of x and y values

```python

x = np.linspace(-3, 3, 50)
y = np.linspace(-3, 3, 50)
X, Y = np.meshgrid(x, y)

# np.linspace(-3, 3, 50) → makes 50 evenly spaced numbers from -3 to 3.
# np.meshgrid(x, y) → creates a grid of all possible (x, y) coordinate pairs

```

##### Step 3: Define the function f(x, y) = 2x³ + 3y

```python

Z = 2 * X**3 + 3 * Y 
```

##### Step 4: Compute partial derivatives

```python

# ∂f/∂x = 6x² constant y,  ∂f/∂y = 3 constant x
df_dx = 6 * X**2 # formula = nx^n-1
df_dy = 3 * np.ones_like(Y) # formula = nx^n-1

# 6 * X**2 gives slope in x-direction for each point.
# Create a new array with the same shape as Y,and fill every element with the number 1
# constant slope everywhere, so we fill with 3’s using np.ones_like(Y).
```

##### Step 5: Plot the surface of f(x, y)

```python

fig = plt.figure(figsize=(10,7)) # width, height
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.8)

# add_subplot(nrows, ncols, index) one big plot 
# 111 means 1 → number of rows of plots1 → number of columns of plots, 1 → this is the first (and only) plot
# plot_surface() draws a smooth colored surface:
# cmap='viridis' adds nice color shading.
# alpha=0.8 makes it slightly transparent

ax.set_title('3D Surface and Gradient: f(x, y) = 2x³ + 3y', fontsize=13)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('f(x, y)')
plt.show()
```
