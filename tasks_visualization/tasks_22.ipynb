##### 22.	Weather Forecast Trend Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample weather data

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Avg_Temperature': [25, 27, 30, 32, 35],
    'Rainfall_mm': [50, 40, 60, 70, 80]
}

df = pd.DataFrame(data)
```

##### Line chart for temperature trend

```python
plt.figure(figsize=(8,5))
plt.plot(df['Month'], df['Avg_Temperature'], marker='o', color='orange')
plt.title("Average Temperature Trend")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
plt.grid(True)
plt.show()

# Temperature increases gradually from January to May → expected seasonal trend
# Rainfall is highest in May → peak rainy month in the dataset
# Line chart clearly shows temperature trend over months
# Bar chart helps compare rainfall month by month
```

##### Bar chart for rainfall

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Month', y='Rainfall_mm', data=df, color='skyblue')
plt.title("Monthly Rainfall")
plt.show()
```
