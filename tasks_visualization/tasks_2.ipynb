##### 2.	Monthly Sales Report Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample monthly sales data

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [20000, 25000, 30000, 28000, 35000, 40000]
}

df = pd.DataFrame(data)
```

##### Line chart for monthly sales trend

```python
plt.figure(figsize=(8,5))
plt.plot(df['Month'], df['Sales'], marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales (â‚¹)")
plt.grid(True)
plt.show()

# Sales show a steady increase from January to June
# Highest sales recorded in June
# Minor dip observed in April
# Charts clearly show monthly sales performance
```

##### Bar chart for monthly sales comparison

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x='Month',
    y='Sales',
    data=df,
    hue='Month',
    palette='pastel',
    dodge=False,
    legend=False
)
plt.title("Monthly Sales Report")
plt.ylabel("Sales (â‚¹)")
plt.show()
```
