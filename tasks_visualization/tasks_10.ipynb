##### 10.	Mobile Usage Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample mobile usage data

```python
data = {
    'App': ['WhatsApp', 'Instagram', 'YouTube', 'Facebook', 'Twitter'],
    'Usage_Minutes': [120, 90, 150, 60, 45]
}

df = pd.DataFrame(data)
```

##### Pie chart for app usage distribution

```python
plt.figure()
plt.pie(df['Usage_Minutes'], labels=df['App'], autopct='%1.1f%%', startangle=90)
plt.title("App Usage Distribution")
plt.show()

# YouTube has the highest usage → most engaging app
# Twitter has the lowest usage → less time spent by users
# Pie chart shows the percentage distribution of app usage clearly
# Bar chart allows easy comparison of usage duration across apps
```

##### Bar chart for usage duration

```python
plt.figure(figsize=(8,5))
sns.barplot(x='App', y='Usage_Minutes', data=df, hue='App', palette='pastel', dodge=False, legend=False)
plt.title("Usage Duration by App")
plt.ylabel("Minutes Used")
plt.show()
```
