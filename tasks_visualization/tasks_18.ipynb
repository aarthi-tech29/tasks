##### 18.	Customer Purchase Behavior

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample customer purchase data

```python
data = {
    'Category': ['Electronics', 'Clothing', 'Home', 'Books', 'Beauty'],
    'Sales': [15000, 12000, 8000, 5000, 4000]
}

df = pd.DataFrame(data)
```

##### Pie chart for purchase distribution by category

```python
plt.figure()
plt.pie(df['Sales'], labels=df['Category'], autopct='%1.1f%%', startangle=90)
plt.title("Purchase Distribution by Category")
plt.show()

# Electronics accounts for the highest sales → most popular category
# Beauty category has the lowest sales → may need promotions or marketing campaigns
# Pie chart shows percentage contribution of each category clearly
# Bar chart helps compare total sales across categories
```

##### Bar chart for total sales by category

```python
plt.figure()
sns.barplot(x='Category', y='Sales', data=df)
plt.title("Total Sales by Category")
plt.show()
```
