##### 32.	Online Course Enrollment Analysis


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample online course enrollment data

```python
data = {
    'Course': ['Python', 'Data Science', 'Web Development', 'Machine Learning'],
    'Enrollments': [300, 450, 200, 350]
}

df = pd.DataFrame(data)
```

##### Pie chart for enrollment distribution by course

```python
plt.figure()
plt.pie(df['Enrollments'], labels=df['Course'], autopct='%1.1f%%', startangle=90)
plt.title("Enrollment Distribution by Course")
plt.show()

# Data Science course has the highest enrollments → most popular course
# Web Development has the lowest enrollments → may need promotions or improved content
# Pie chart shows the percentage share of enrollments for each course
# Bar chart helps compare enrollment counts across courses clearly
```

##### Bar chart for enrollment comparison

```python
plt.figure()
sns.barplot(x='Course', y='Enrollments', data=df)
plt.title("Enrollment Count by Course")
plt.show()
```
