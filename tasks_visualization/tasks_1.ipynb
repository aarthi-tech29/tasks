##### 1.	Student Marks Visualization (Bar, Pie)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample student marks data

```python
data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Computer'],
    'Marks': [85, 90, 78, 70, 95]
}

df = pd.DataFrame(data)
```

##### Bar chart for student marks

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x='Subject',
    y='Marks',
    data=df,
    hue='Subject',
    palette='pastel',
    dodge=False,
    legend=False
)
plt.title("Student Marks by Subject")
plt.ylabel("Marks")
plt.ylim(0, 100)
plt.show()

# Computer has the highest marks → strongest subject
# History has the lowest marks → needs improvement
# Bar chart clearly compares marks across subjects
# Pie chart shows percentage contribution of each subject
```

##### Pie chart for marks distribution

```python
plt.figure()
plt.pie(
    df['Marks'],
    labels=df['Subject'],
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Marks Distribution by Subject")
plt.show()
```
