##### 46.	Energy Consumption Prediction Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample energy data

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Actual_Consumption': [320, 300, 280, 260, 290],
    'Predicted_Consumption': [330, 310, 270, 255, 295]
}

df = pd.DataFrame(data)
```

##### Line chart: Actual vs Predicted

```python
plt.figure()
plt.plot(df['Month'], df['Actual_Consumption'], label='Actual')
plt.plot(df['Month'], df['Predicted_Consumption'], label='Predicted')
plt.title("Energy Consumption: Actual vs Predicted")
plt.xlabel("Month")
plt.ylabel("Energy (kWh)")
plt.legend()
plt.show()

# Predicted values closely follow actual consumption
# Energy usage decreases during moderate months
# Small differences indicate accurate prediction
```
