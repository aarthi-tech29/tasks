##### 44.	Marketing Campaign Performance Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample campaign data

```python
data = {
    'Campaign': ['Email', 'Social Media', 'Ads', 'Email', 'Ads'],
    'Conversions': [120, 200, 150, 130, 170],
    'Cost': [30000, 40000, 50000, 32000, 48000]
}

df = pd.DataFrame(data)
```

##### Bar chart

```python
plt.figure()
sns.barplot(x='Campaign', y='Conversions', data=df)
plt.title("Conversions by Marketing Campaign")
plt.show()


# Social Media campaigns show higher conversions
# Ad campaigns have higher costs
# Email campaigns are cost-effective
```

##### Box plot

```python
plt.figure()
sns.boxplot(x='Campaign', y='Cost', data=df)
plt.title("Campaign Cost Distribution")
plt.show()
```
