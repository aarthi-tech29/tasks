##### 26.	Social Media Engagement Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample social media engagement data

```python
data = {
    'Platform': ['Facebook', 'Instagram', 'Twitter', 'LinkedIn'],
    'Engagements': [1200, 1800, 900, 600]
}

df = pd.DataFrame(data)
```

##### Pie chart for engagement distribution by platform

```python
plt.figure()
plt.pie(df['Engagements'], labels=df['Platform'], autopct='%1.1f%%', startangle=90)
plt.title("Engagement Distribution by Platform")
plt.show()

# Instagram has the highest engagement → most active audience
# LinkedIn has the lowest engagement → may need different content strategy
# Pie chart shows percentage share of engagement by platform
# Bar chart clearly compares engagement counts across platforms
```

##### Bar chart for engagement comparison

```python
plt.figure()
sns.barplot(x='Platform', y='Engagements', data=df)
plt.title("Engagement Count by Platform")
plt.show()
```
