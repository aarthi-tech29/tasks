##### 13.	Age Distribution Histogram

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample age data

```python
ages = [15, 16, 15, 14, 16, 15, 17, 16, 14, 15, 17, 18, 16, 15, 14, 17, 18, 16, 15, 14]
```

##### Histogram

```python
plt.figure(figsize=(8,5))
plt.hist(ages, bins=5, color='skyblue', edgecolor='black')
plt.title("Age Distribution Histogram")
plt.xlabel("Age")
plt.ylabel("Number of Individuals")
plt.show()

# Most individuals are aged 15-16 â†’ majority in this age range
# Histogram clearly shows frequency of each age group
# Density plot provides a smooth visualization of age distribution
```

##### Density plot using Seaborn

```python
plt.figure(figsize=(8,5))
sns.kdeplot(ages, fill=True, color='orange') 
plt.title("Age Density Plot")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show()

```
