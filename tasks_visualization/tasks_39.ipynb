##### 39.	E-commerce Sales Forecasting Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample e-commerce sales data

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Actual_Sales': [50000, 60000, 75000, 90000, 105000],
    'Forecasted_Sales': [52000, 62000, 73000, 88000, 108000]
}

df = pd.DataFrame(data)
```

##### Line chart: Actual vs Forecasted Sales

```python
plt.figure()
plt.plot(df['Month'], df['Actual_Sales'], label='Actual Sales')
plt.plot(df['Month'], df['Forecasted_Sales'], label='Forecasted Sales')
plt.title("E-commerce Sales: Actual vs Forecast")
plt.xlabel("Month")
plt.ylabel("Sales Amount")
plt.legend()
plt.show()

# Sales show a steady upward trend
# Forecasted sales closely match actual sales
# Minor variations indicate realistic predictions
```

##### Bar chart

```python
plt.figure()
sns.barplot(x='Month', y='Actual_Sales', data=df)
plt.title("Monthly E-commerce Sales")
plt.show()
```
