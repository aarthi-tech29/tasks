##### 17.	Sales vs Profit Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample sales vs profit data

```python
data = {
    'Category': ['Electronics', 'Clothing', 'Home', 'Books', 'Beauty'],
    'Sales': [15000, 12000, 8000, 5000, 4000],
    'Profit': [5000, 4000, 2000, 1000, 800]
}

df = pd.DataFrame(data)
```

##### Scatter plot for Sales vs Profit

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Sales', y='Profit', data=df, hue='Category', s=100)
plt.title("Sales vs Profit by Category")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.grid(True)
plt.show()

# Electronics has the highest sales and profit → most profitable category
# Beauty has the lowest sales and profit → may need cost optimization or marketing
# Scatter plot shows the relationship between sales and profit clearly
# Bar chart helps compare profit across categories
```

##### Bar chart for profit by category

```python
plt.figure()
sns.barplot(x='Category', y='Profit', data=df)
plt.title("Profit by Category")
plt.show()
```
