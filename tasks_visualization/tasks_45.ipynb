##### 45.	Startup Growth Analytics

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample startup data

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Users': [500, 800, 1200, 1800, 2500],
    'Revenue': [20000, 35000, 50000, 75000, 100000],
    'Expenses': [15000, 20000, 30000, 40000, 55000]
}

df = pd.DataFrame(data)
df['Profit'] = df['Revenue'] - df['Expenses']
print(df)
```

##### Line chart

```python
plt.figure()
plt.plot(df['Month'], df['Users'])
plt.title("Startup User Growth")
plt.xlabel("Month")
plt.ylabel("Users")
plt.show()

# User base shows steady growth
# Revenue increases faster than expenses
# Profit rises month by month, indicating healthy startup growth
```

##### Bar chart

```python
plt.figure()
sns.barplot(x='Month', y='Profit', data=df)
plt.title("Monthly Profit")
plt.show()
```
