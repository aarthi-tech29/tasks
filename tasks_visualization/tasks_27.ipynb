##### 27.	E-commerce Category-wise Sales

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample e-commerce sales data

```python
data = {
    'Category': ['Electronics', 'Clothing', 'Home Appliances', 'Books', 'Beauty'],
    'Revenue': [15000, 12000, 8000, 5000, 4000]
}

df = pd.DataFrame(data)
```

##### Pie chart for revenue distribution by category

```python
plt.figure()
plt.pie(df['Revenue'], labels=df['Category'], autopct='%1.1f%%', startangle=90)
plt.title("Revenue Distribution by Category")
plt.show()

# Electronics contributes the highest revenue → most profitable category
# Beauty category has the lowest revenue → may require marketing promotions
# Pie chart shows percentage contribution of each category clearly
# Bar chart helps compare revenue across categories side by side
```

##### Bar chart for revenue comparison

```python
plt.figure()
sns.barplot(x='Category', y='Revenue', data=df)
plt.title("Revenue by Category")
plt.show()
```
