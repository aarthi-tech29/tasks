##### 15.	Simple E-commerce Product Analysis


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample product sales data

```python
data = {
    'Product': ['Laptop', 'Phone', 'Headphones', 'Camera', 'Smartwatch'],
    'Sales': [15000, 12000, 5000, 7000, 4000]
}

df = pd.DataFrame(data)
```

##### Pie chart for sales distribution by product

```python
plt.figure()
plt.pie(df['Sales'], labels=df['Product'], autopct='%1.1f%%', startangle=90)
plt.title("Sales Distribution by Product")
plt.show()

# Laptop has the highest sales → most popular product
# Smartwatch has the lowest sales → may need marketing or discount strategies
# Pie chart shows percentage contribution of each product clearly
# Bar chart allows easy comparison of product sales
```

##### Bar chart for sales by product

```python
plt.figure()
sns.barplot(x='Product', y='Sales', data=df, hue='Product', palette='pastel')
plt.title("Sales by Product")
plt.show()
```
