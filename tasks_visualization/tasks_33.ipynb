##### 33.	Restaurant Sales Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample restaurant sales data

```python
data = {
    'Category': ['Fast Food', 'Beverage', 'Dessert', 'Main Course'],
    'Revenue': [5000, 2000, 1500, 4500]
}

df = pd.DataFrame(data)
```

##### Pie chart for revenue distribution by category

```python
plt.figure()
plt.pie(df['Revenue'], labels=df['Category'], autopct='%1.1f%%', startangle=90)
plt.title("Revenue Distribution by Category")
plt.show()
```

##### Bar chart for revenue comparison

```python
plt.figure()
sns.barplot(x='Category', y='Revenue', data=df)
plt.title("Revenue by Category")
plt.show()

# Fast Food contributes the highest revenue  main source of income
# Desserts have the lowest revenue  may need promotions or new items
# Pie chart shows percentage contribution of each category
# Bar chart helps compare revenue across categories clearly
```
