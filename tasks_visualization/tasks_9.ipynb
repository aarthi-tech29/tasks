##### 9.	Exam Result Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample exam result data


```python
data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Computer'],
    'Average_Marks': [75, 82, 70, 65, 88]
}

df = pd.DataFrame(data)
```

##### Bar chart for subject-wise average marks 

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x='Subject',
    y='Average_Marks',
    data=df,
    hue='Subject',
    palette='pastel',
    dodge=False,
    legend=False
)
plt.title("Subject-wise Average Marks")
plt.ylabel("Marks")
plt.show()

# Computer has the highest average marks → strong subject performance
# History has the lowest average marks → may need extra support
# Majority of students passed the exam, indicating good overall performance
# Bar chart highlights subject-wise strengths and weaknesses clearly
```

##### Pass vs Fail pie chart

```python
result_data = {'Pass': 45, 'Fail': 5}
result_df = pd.DataFrame(list(result_data.items()), columns=['Result', 'Count'])
print(result_df)
plt.figure()
plt.pie(result_df['Count'], labels=result_df['Result'],
        autopct='%1.1f%%', startangle=90)
plt.title("Pass vs Fail Distribution")
plt.show()
```
