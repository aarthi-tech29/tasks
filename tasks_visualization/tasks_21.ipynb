##### 21.	COVID-19 Data Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample COVID-19 data

```python
data = {
    'Date': pd.date_range(start='2025-12-01', periods=5, freq='D'),
    'Confirmed': [1000, 1200, 1500, 1700, 2000],
    'Recovered': [800, 900, 1100, 1300, 1600],
    'Deaths': [50, 60, 70, 80, 90]
}

df = pd.DataFrame(data)
```

##### Line chart for COVID-19 trends

```python
plt.figure(figsize=(8,5))
plt.plot(df['Date'], df['Confirmed'], marker='o', label='Confirmed')
plt.plot(df['Date'], df['Recovered'], marker='o', label='Recovered')
plt.plot(df['Date'], df['Deaths'], marker='o', label='Deaths')
plt.title("COVID-19 Cases Trend")
plt.xlabel("Date")
plt.ylabel("Number of Cases")
plt.legend()
plt.grid(True)
plt.show()


# Confirmed cases are increasing steadily  monitor spread
# Recovery rate is high  effective treatment
# Deaths are comparatively low  positive sign
# Line chart shows trend over time clearly
# Pie chart shows current distribution of active, recovered, and death cases
```

##### Pie chart for latest day case distribution

```python
latest_data = df.iloc[-1]
labels = ['Recovered', 'Deaths', 'Active']
sizes = [latest_data['Recovered'], latest_data['Deaths'], latest_data['Confirmed'] - latest_data['Recovered'] - latest_data['Deaths']]

plt.figure()
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title("COVID-19 Case Distribution (Latest Day)")
plt.show()
```
