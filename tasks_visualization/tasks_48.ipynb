##### 48.	Bank Fraud Detection Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample banking transaction data

```python
data = {
    'Amount': [500, 2000, 10000, 1500, 25000],
    'Fraud': ['No', 'No', 'Yes', 'No', 'Yes']
}

df = pd.DataFrame(data)
```

##### Bar chart

```python
plt.figure()
sns.countplot(x='Fraud', data=df)
plt.title("Fraud vs Non-Fraud Transactions")
plt.show()

# Fraud transactions usually involve higher amounts
# Box plot shows clear difference between fraud and non-fraud transactions
# Fewer fraud cases compared to normal transactions
```

##### Box plot

```python
plt.figure()
sns.boxplot(x='Fraud', y='Amount', data=df)
plt.title("Transaction Amount by Fraud Status")
plt.show()
```
