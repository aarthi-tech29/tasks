##### 36.	Business Intelligence Dashboard

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample business data

```python
data = {
    'Region': ['North', 'South', 'East', 'West'],
    'Sales': [250000, 180000, 220000, 200000],
    'Profit': [50000, 30000, 45000, 40000]
}

df = pd.DataFrame(data)
```

##### Bar chart

```python
plt.figure()
sns.barplot(x='Region', y='Sales', data=df)
plt.title("Sales by Region")
plt.show()

# North region has the highest sales
# Profit distribution shows variation across regions
# BI dashboard highlights strong and weak business areas
```

##### Box plot

```python
plt.figure()
sns.boxplot(y='Profit', data=df)
plt.title("Profit Distribution")
plt.show()
```
