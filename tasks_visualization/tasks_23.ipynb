##### 23.	IPL Match Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample IPL team win data

```python
data = {
    'Team': ['Mumbai Indians', 'Chennai Super Kings', 'RCB', 'Delhi Capitals'],
    'Wins': [9, 8, 7, 6]
}

df = pd.DataFrame(data)
```

##### Pie chart for win percentage

```python
plt.figure()
plt.pie(df['Wins'], labels=df['Team'], autopct='%1.1f%%', startangle=90)
plt.title("Team Win Percentage")
plt.show()

# Mumbai Indians have the highest wins → most successful team in the dataset
# Delhi Capitals have the lowest wins → needs improvement
# Pie chart shows win percentage contribution of each team clearly
# Bar chart allows easy comparison of team performance
```

##### Bar chart for wins comparison

```python
plt.figure()
sns.barplot(x='Team', y='Wins', data=df)
plt.title("Wins by Team")
plt.show()
```
