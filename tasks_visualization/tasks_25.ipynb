##### 25.	Hospital Patient Data Analytics

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample patient data

```python
data = {
    'Department': ['Cardiology', 'Neurology', 'Orthopedics', 'Pediatrics', 'General'],
    'Patients': [120, 80, 90, 70, 150]
}

df = pd.DataFrame(data)
```

##### Pie chart for patient distribution by department

```python
plt.figure()
plt.pie(df['Patients'], labels=df['Department'], autopct='%1.1f%%', startangle=90)
plt.title("Patient Distribution by Department")
plt.show()

# General department has the highest patient count → high demand
# Pediatrics has the lowest patient count → specialized care
# Pie chart shows percentage distribution of patients by department
# Bar chart helps compare patient counts clearly across departments
```

##### Bar chart for patient count comparison

```python
plt.figure()
sns.barplot(x='Department', y='Patients', data=df)
plt.title("Patient Count by Department")
plt.show()
```
