##### 3.	Employee Salary Distribution

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample salary data

```python
salaries = [25000, 30000, 28000, 35000, 40000,
            45000, 50000, 32000, 37000, 60000]
```

##### Histogram for salary distribution

```python
plt.figure(figsize=(8,5))
plt.hist(salaries, bins=6, edgecolor='black')
plt.title("Employee Salary Distribution")
plt.xlabel("Salary (₹)")
plt.ylabel("Number of Employees")
plt.show()

# Most employees earn between ₹30,000 and ₹45,000
# Few employees earn very high salaries → senior-level roles
# Histogram shows how salaries are distributed
# Density plot highlights common salary ranges
```

##### Density plot

```python
plt.figure(figsize=(8,5))
sns.kdeplot(salaries, fill=True)
plt.title("Salary Density Plot")
plt.xlabel("Salary (₹)")
plt.ylabel("Density")
plt.show()
```
