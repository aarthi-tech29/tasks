##### 35.	Product Review Rating Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample product rating data

```python
data = {
    'Product': ['Product A', 'Product B', 'Product C', 'Product D'],
    'Rating': [4.5, 3.8, 4.2, 2.9]
}

df = pd.DataFrame(data)
```

##### Bar chart

```python
plt.figure()
sns.barplot(x='Product', y='Rating', data=df)
plt.title("Average Product Ratings")
plt.show()

# Product A has the highest customer rating
# Product D has lower satisfaction
# Box plot shows variation and possible outliers in ratings
```

##### Box plot

```python
plt.figure()
sns.boxplot(y='Rating', data=df)
plt.title("Rating Distribution")
plt.show()
```
