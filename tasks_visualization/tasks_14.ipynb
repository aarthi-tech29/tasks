##### 14.	College Department Strength Chart


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample department data

```python
data = {
    'Department': ['Computer Science', 'Mechanical', 'Electrical', 'Civil', 'Biotech'],
    'Students': [120, 100, 80, 60, 50]
}

df = pd.DataFrame(data)
```

##### Pie chart for department-wise student distribution

```python
plt.figure()
plt.pie(df['Students'], labels=df['Department'], autopct='%1.1f%%', startangle=90)
plt.title("Student Distribution by Department")
plt.show()

# Computer Science has the highest number of students → largest department
# Biotech has the lowest number of students → smallest department
# Pie chart shows the percentage share of students in each department clearly
# Bar chart allows easy comparison of student counts across departments
```

##### Bar chart for student count by department

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Department', y='Students', data=df, hue='Department', palette='pastel', dodge=False, legend=False)
plt.title("Student Count by Department")
plt.ylabel("Number of Students")
plt.show()
```
