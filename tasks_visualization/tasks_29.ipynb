##### 29.	Electricity Consumption Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample electricity consumption data

```python
data = {
    'Category': ['Residential', 'Commercial', 'Industrial'],
    'Consumption_kWh': [1200, 1800, 2500]
}

df = pd.DataFrame(data)
```

##### Pie chart for consumption distribution by category

```python
plt.figure()
plt.pie(df['Consumption_kWh'], labels=df['Category'], autopct='%1.1f%%', startangle=90)
plt.title("Electricity Consumption Distribution")
plt.show()

# Industrial category consumes the most electricity → largest share of usage
# Residential category consumes the least → lower share, but important for planning
# Pie chart shows percentage contribution of each category clearly
# Bar chart helps compare consumption across categories
```

##### Bar chart for consumption comparison

```python
plt.figure()
sns.barplot(x='Category', y='Consumption_kWh', data=df)
plt.title("Electricity Consumption by Category")
plt.show()
```
