##### 20.	Bank Loan Approval Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample loan data

```python
data = {
    'Loan_Status': ['Approved', 'Rejected'],
    'Count': [200, 80]
}

df = pd.DataFrame(data)
```

##### Pie chart for loan approval vs rejection

```python
plt.figure()
plt.pie(df['Count'], labels=df['Loan_Status'], autopct='%1.1f%%', startangle=90)
plt.title("Loan Approval vs Rejection")
plt.show()

# Approved loans are higher than rejected â†’ majority of applicants get approved
# Pie chart shows approval percentage clearly
# Bar chart helps compare approved vs rejected loan counts easily
# Banks can focus on patterns among rejected applicants to improve approval rates
```

##### Bar chart for loan status comparison

```python
plt.figure()
sns.barplot(x='Loan_Status', y='Count', data=df)
plt.title("Loan Status Count Comparison")
plt.show()
```
