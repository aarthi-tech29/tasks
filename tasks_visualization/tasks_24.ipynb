##### 24.	Crime Rate Analysis (Year-wise)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample year-wise crime data

```python
data = {
    'Year': [2021, 2022, 2023, 2024, 2025],
    'Crime_Count': [500, 450, 480, 430, 400]
}

df = pd.DataFrame(data)
```

##### Line chart for crime trend over years

```python
plt.figure(figsize=(8,5))
plt.plot(df['Year'], df['Crime_Count'], marker='o')
plt.title("Year-wise Crime Trend")
plt.xlabel("Year")
plt.ylabel("Number of Cases")
plt.grid(True)
plt.show()

# Crime cases decreased gradually from 2021 to 2025 â†’ positive trend in law enforcement
# Line chart shows the year-wise trend clearly, helping identify rising or falling patterns
# Bar chart helps compare crime counts across different years
```

##### Bar chart for crime count by year

```python
plt.figure()
sns.barplot(x='Year', y='Crime_Count', data=df)
plt.title("Crime Count by Year")
plt.show()
```
