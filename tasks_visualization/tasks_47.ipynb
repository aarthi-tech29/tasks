##### 47.	Sports Analytics Dashboard

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample sports data

```python
data = {
    'Player': ['Player A', 'Player B', 'Player C', 'Player D'],
    'Runs': [450, 520, 380, 610]
}

df = pd.DataFrame(data)
```

##### Bar chart

```python
plt.figure()
sns.barplot(x='Player', y='Runs', data=df)
plt.title("Player Performance (Runs)")
plt.show()

# Player D has the highest performance
# Performance varies significantly among players
# Box plot shows score spread and median clearly
```

##### Box plot

```python
plt.figure()
sns.boxplot(y='Runs', data=df)
plt.title("Distribution of Player Runs")
plt.show()
```
