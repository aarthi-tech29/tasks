##### 7.	Attendance Percentage Graph

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample attendance data

```python
data = {
    'Student': ['A', 'B', 'C', 'D', 'E'],
    'Attendance_Percentage': [92, 85, 78, 88, 95]
}

df = pd.DataFrame(data)
```

##### Bar chart for attendance percentage

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x='Student',
    y='Attendance_Percentage',
    data=df,
    hue='Student',
    palette='pastel',
    dodge=False,
    legend=False
)
plt.title("Attendance Percentage by Student")
plt.ylabel("Attendance (%)")
plt.ylim(0, 100)
plt.show()

# Student E has the highest attendance → very regular
# Student C has the lowest attendance → may need monitoring
# Bar chart clearly compares attendance percentages among students
```
