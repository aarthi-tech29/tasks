##### 41.	Healthcare Analytics Dashboard

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample healthcare data

```python
data = {
    'Disease': ['Diabetes', 'Heart', 'Cancer', 'Diabetes', 'Heart'],
    'Patients': [120, 90, 60, 110, 100],
    'Cost': [30000, 50000, 70000, 28000, 52000]
}

df = pd.DataFrame(data)
```

#####  Bar chart

```python
plt.figure()
sns.barplot(x='Disease', y='Patients', data=df)
plt.title("Number of Patients by Disease")
plt.show()
```

##### Box plot

```python
plt.figure()
sns.boxplot(x='Disease', y='Cost', data=df)
plt.title("Treatment Cost by Disease")
plt.show()
```
