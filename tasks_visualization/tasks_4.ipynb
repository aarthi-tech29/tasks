##### 4.	Weather Report Visualization


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample weather data

```python
data = {
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Temperature': [30, 32, 31, 33, 34],
    'Rainfall': [5, 0, 10, 2, 0]
}

df = pd.DataFrame(data)
```

##### Line chart for temperature trend

```python
plt.figure(figsize=(8,5))
plt.plot(df['Day'], df['Temperature'], marker='o')
plt.title("Daily Temperature Trend")
plt.xlabel("Day")
plt.ylabel("Temperature (Â°C)")
plt.grid(True)
plt.show()

# Temperature shows a gradual increase over the days
# Thursday and Friday are the hottest days
# Rainfall is highest on Wednesday

```

##### Bar chart for rainfall

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x='Day',
    y='Rainfall',
    data=df,
    hue='Day',
    palette='pastel',
    dodge=False,
    legend=False
)
plt.title("Daily Rainfall Report")
plt.ylabel("Rainfall (mm)")
plt.show()
```
