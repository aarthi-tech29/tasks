##### 5.	Simple Line Chart – Daily Expenses

```python
import pandas as pd
import matplotlib.pyplot as plt
```

##### Sample daily expense data

```python
data = {
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'],
    'Expenses': [250, 300, 200, 350, 400, 500, 450]
}

df = pd.DataFrame(data)
```

##### Line chart for daily expenses

```python
plt.figure(figsize=(8,5))
plt.plot(df['Day'], df['Expenses'], marker='o')
plt.title("Daily Expenses Trend")
plt.xlabel("Day")
plt.ylabel("Expenses (₹)")
plt.grid(True)
plt.show()
```
