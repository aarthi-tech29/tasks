##### 11.	Fuel Price Trend Graph


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample fuel price data

```python
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Petrol': [105, 107, 110, 112, 115],
    'Diesel': [95, 96, 98, 100, 102]
}

df = pd.DataFrame(data)
```

##### Line chart for petrol and diesel price trends

```python
plt.figure(figsize=(10,6))
plt.plot(df['Month'], df['Petrol'], marker='o', label='Petrol', color='orange')
plt.plot(df['Month'], df['Diesel'], marker='o', label='Diesel', color='green')
plt.title("Fuel Price Trend Over Months")
plt.xlabel("Month")
plt.ylabel("Price (₹)")
plt.legend()
plt.grid(True)
plt.show()

# Petrol and diesel prices are increasing from January to May → upward trend
# Line chart clearly shows monthly price trends for each fuel type
# Bar chart allows easy month-wise and fuel-wise comparison
```

##### Bar chart for comparison

```python
df_melted = df.melt(id_vars='Month', var_name='Fuel', value_name='Price')
plt.figure(figsize=(10,6))
sns.barplot(x='Month', y='Price', hue='Fuel', data=df_melted, palette='pastel')
plt.title("Fuel Price Comparison")
plt.ylabel("Price (₹)")
plt.show()
```
