##### 34.	Gender Ratio Analytics

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample gender data

```python
data = {
    'Gender': ['Male', 'Female'],
    'Count': [520, 480]
}

df = pd.DataFrame(data)
```

##### Pie chart

```python
plt.figure()
plt.pie(df['Count'], labels=df['Gender'], autopct='%1.1f%%')
plt.title("Gender Ratio")
plt.show()

# Male and female counts are almost balanced
# Pie chart clearly shows gender percentage
# Bar chart helps compare gender counts easily
```

##### Bar chart

```python
plt.figure()
sns.barplot(x='Gender', y='Count', data=df)
plt.title("Gender Count Comparison")
plt.show()
```
