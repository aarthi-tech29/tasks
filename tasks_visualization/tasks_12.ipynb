##### 12.	Simple Survey Result Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample survey data

```python
data = {
    'Response': ['Yes', 'No', 'Maybe'],
    'Count': [60, 25, 15]
}

df = pd.DataFrame(data)
```

##### Pie chart for survey results

```python
plt.figure()
plt.pie(df['Count'], labels=df['Response'], autopct='%1.1f%%', startangle=90)
plt.title("Survey Result Distribution")
plt.show()

# Majority of respondents answered Yes → positive sentiment
# Maybe has the lowest count → uncertain responses are few
# Pie chart shows percentage distribution of survey responses clearly
# Bar chart allows easy comparison of response counts
```

##### Bar chart for survey results

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Response', y='Count', data=df, hue='Response', palette='pastel', dodge=False, legend=False)
plt.title("Survey Result Count")
plt.ylabel("Number of Responses")
plt.show()
```
