##### 28.	Stock Price Trend Analysis


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample stock price data

```python
data = {
    'Date': pd.date_range(start='2025-12-01', periods=5, freq='D'),
    'Close_Price': [150, 155, 152, 158, 160],
    'Volume': [1000, 1200, 1100, 1500, 1300]
}

df = pd.DataFrame(data)
```

##### Line chart for stock price trend

```python
plt.figure(figsize=(8,5))
plt.plot(df['Date'], df['Close_Price'], marker='o')
plt.title("Stock Price Trend")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.grid(True)
plt.show()

# Stock price shows an upward trend over the observed period → positive growth
# Daily volume peaks on the 4th day → highest trading activity
# Line chart shows trend clearly over time
# Bar chart shows trading volume and helps identify high-activity days
```

##### Bar chart for traded volume

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Date', y='Volume', data=df)
plt.title("Daily Traded Volume")
plt.show()
```
