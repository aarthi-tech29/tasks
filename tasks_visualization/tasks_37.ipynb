##### 37.	Student Placement Analytics

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample placement data

```python
data = {
    'Department': ['CSE', 'ECE', 'EEE', 'CSE', 'ECE'],
    'CGPA': [8.5, 7.8, 7.0, 9.1, 8.0],
    'Salary_LPA': [6, 4.5, 3.8, 8, 5.2]
}

df = pd.DataFrame(data)
```

##### Bar chart

```python
plt.figure()
sns.barplot(x='Department', y='Salary_LPA', data=df)
plt.title("Average Salary by Department")
plt.show()

# Higher CGPA students receive better salary packages
# CSE department shows higher average salary
# Scatter plot shows a positive relationship between CGPA and salary
```

##### Scatter plot

```python
plt.figure()
sns.scatterplot(x='CGPA', y='Salary_LPA', data=df)
plt.title("CGPA vs Salary Package")
plt.show()
```
