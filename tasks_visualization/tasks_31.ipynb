##### 31.	Transport Usage Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample transport usage data

```python
data = {
    'Mode': ['Bus', 'Train', 'Car', 'Bike', 'Metro'],
    'Usage': [300, 450, 500, 150, 350]
}

df = pd.DataFrame(data)
```

##### Pie chart for usage distribution by transport mode

```python
plt.figure()
plt.pie(df['Usage'], labels=df['Mode'], autopct='%1.1f%%', startangle=90)
plt.title("Transport Usage Distribution")
plt.show()

# Car is the most used mode of transport → primary choice for commuters
# Bike has the lowest usage → may indicate need for better bike lanes or promotions
# Pie chart shows percentage contribution of each mode
# Bar chart clearly compares usage counts across different transport modes
```

##### Bar chart for usage comparison

```python
plt.figure()
sns.barplot(x='Mode', y='Usage', data=df)
plt.title("Usage Count by Transport Mode")
plt.show()
```
