##### 49.	Traffic Flow Analytics System

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample traffic data

```python
data = {
    'Hour': ['6AM', '9AM', '12PM', '3PM', '6PM'],
    'Vehicle_Count': [120, 450, 300, 280, 500]
}

df = pd.DataFrame(data)
```

##### Line chart

```python
plt.figure()
plt.plot(df['Hour'], df['Vehicle_Count'])
plt.title("Traffic Flow During the Day")
plt.xlabel("Time")
plt.ylabel("Vehicle Count")
plt.show()

# Traffic peaks during 9 AM and 6 PM
# Noon and afternoon have moderate traffic
# Line chart shows daily traffic trend clearly
```

##### Bar chart

```python
plt.figure()
sns.barplot(x='Hour', y='Vehicle_Count', data=df)
plt.title("Vehicle Count by Time")
plt.show()
```
