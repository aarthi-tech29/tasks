##### 42.	Customer Segmentation Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample customer data

```python
data = {
    'Income': [15000, 30000, 45000, 60000, 75000],
    'SpendingScore': [20, 40, 60, 80, 90],
    'Segment': ['Low', 'Medium', 'Medium', 'High', 'High']
}

df = pd.DataFrame(data)

# Low Segment → Low income, low spending
# Medium Segment → Moderate income and spending
# High Segment → High income, high spending
```

##### Scatter plot

```python
plt.figure()
sns.scatterplot(x='Income', y='SpendingScore', hue='Segment', data=df)
plt.title("Customer Segmentation based on Income and Spending")
plt.show()
```

##### Box plot

```python
plt.figure()
sns.boxplot(x='Segment', y='SpendingScore', data=df)
plt.title("Spending Score by Customer Segment")
plt.show()
```
