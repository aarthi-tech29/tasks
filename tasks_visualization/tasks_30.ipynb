##### 30.	College Result Analytics Dashboard

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample student result data

```python
data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Computer'],
    'Average_Marks': [78, 85, 72, 65, 90]
}

df = pd.DataFrame(data)
print(df)
```

##### Pie chart for pass vs fail (example)

```python
pass_fail_data = {'Pass': 45, 'Fail': 5}
pf_df = pd.DataFrame(list(pass_fail_data.items()), columns=['Status', 'Count'])
print(pf_df)

plt.figure()
plt.pie(pf_df['Count'], labels=pf_df['Status'], autopct='%1.1f%%', startangle=90)
plt.title("Pass vs Fail Percentage")
plt.show()

# Computer has the highest average marks → strong subject performance
# History has the lowest average marks → may need extra attention or tutoring
# Pass vs Fail pie chart → majority of students are passing → overall good performance
# Bar chart clearly shows which subjects students excel in or struggle with
```

#####  Bar chart for average marks by subject

```python
plt.figure()
sns.barplot(x='Subject', y='Average_Marks', data=df)
plt.title("Average Marks by Subject")
plt.show()
```
