##### 19.	Company Employee Analytics Dashboard

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample employee data

```python
data = {
    'Department': ['HR', 'IT', 'Sales', 'Marketing', 'Finance'],
    'Employees': [25, 50, 40, 30, 20]
}

df = pd.DataFrame(data)
```

##### Pie chart for department-wise employee distribution

```python
plt.figure()
plt.pie(df['Employees'], labels=df['Department'], autopct='%1.1f%%', startangle=90)
plt.title("Employee Distribution by Department")
plt.show()

# IT department has the highest number of employees → core team of the company
# Finance has the lowest number of employees → smaller department
# Pie chart shows department-wise employee percentage clearly
# Bar chart helps compare employee count across departments
```

##### Bar chart for employee count by department

```python
plt.figure()
sns.barplot(x='Department', y='Employees', data=df)
plt.title("Employee Count by Department")
plt.show()
```
