##### 6.	Product Price Comparison Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample product price data

```python
data = {
    'Product': ['Laptop', 'Phone', 'Tablet', 'Headphones', 'Smartwatch'],
    'Price': [60000, 30000, 25000, 5000, 12000]
}

df = pd.DataFrame(data)
```

##### Bar chart for product price comparison

```python
plt.figure(figsize=(8,5))
sns.barplot(
    x='Product',
    y='Price',
    data=df,
    hue='Product',
    palette='pastel',
    dodge=False,
    legend=False
)
plt.title("Product Price Comparison")
plt.ylabel("Price (â‚¹)")
plt.show()

# Laptop is the most expensive product
# Headphones are the least expensive product
# Bar chart clearly highlights price differences across products
```
