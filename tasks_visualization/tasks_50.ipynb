```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample student data

```python
data = {
    'Name': ['Arun', 'Divya', 'Karthik', 'Meena'],
    'Department': ['CS', 'ECE', 'CS', 'ME'],
    'Attendance': [85, 78, 92, 70],
    'Marks': [88, 75, 91, 68],
    'Gender': ['Male', 'Male', 'Male', 'Female']
}

df = pd.DataFrame(data)
```

##### Bar chart - Department Strength

```python
plt.figure(figsize=(6,4))
sns.countplot(
    x='Department', 
    data=df, 
    hue='Department',   # Assign x to hue
    palette='pastel', 
    dodge=False,        # ensures bars don't split
    legend=False        # remove legend
)
plt.title("Department Strength")
plt.show()
```

##### Attendance vs Marks Scatter Plot

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Attendance', y='Marks', hue='Department', data=df, s=100, palette='bright')
plt.title("Attendance vs Marks")
plt.show()

# Department Strength: CS has more students → can plan resources accordingly
# Attendance vs Marks: Students with higher attendance generally score better
# Gender Ratio: Balanced gender distribution → supports equality policies
```

##### Pie chart - Gender Ratio Example (if gender column exists)

```python
gender_counts = df['Gender'].value_counts()
plt.figure()
plt.pie(
    gender_counts,
    labels=gender_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Gender Ratio")
plt.show()
```

##### Gender count bar chart

```python
plt.figure(figsize=(6,4))
sns.countplot(
    x='Gender',
    data=df,
    hue='Gender',  
    palette='pastel',
    dodge=False,
    legend=False
)
plt.title("Gender Count Comparison")
plt.show()
```
