##### 16.	Student Performance Analytics (All subjects)


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Sample student performance data

```python
data = {
    'Subject': ['Math', 'Science', 'English', 'History', 'Computer'],
    'Average_Marks': [78, 85, 72, 65, 90]
}

df = pd.DataFrame(data)
```

##### Bar chart for average marks by subject

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Subject', y='Average_Marks', data=df,hue='Subject', palette='pastel')
plt.title("Average Marks by Subject")
plt.ylabel("Marks")
plt.show()

# Computer has the highest average marks → students excel in this subject
# History has the lowest average marks → may need extra attention or coaching
# Bar chart shows subject-wise performance clearly
# Pie chart shows the overall grade distribution in the class
```

##### Pie chart for grade distribution (sample data)

```python
grades = {'A': 10, 'B': 15, 'C': 5}
grades_df = pd.DataFrame(list(grades.items()), columns=['Grade', 'Count'])
print(grades_df)
plt.figure()
plt.pie(grades_df['Count'], labels=grades_df['Grade'], autopct='%1.1f%%', startangle=90)
plt.title("Grade Distribution")
plt.show()
```
