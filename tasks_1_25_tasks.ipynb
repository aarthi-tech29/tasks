##### 12. Check whether a year is a leap year

```python
year = 2024

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap Year")
else:
    print("Not a Leap Year")

# 2024 % 4 = 0 
# 2024 % 100 = 24 → not equal to 0
```

##### 17. Assign grades based on marks using if-elif-else

```python
# Given marks
marks = 78

# Assign grades
if marks >= 90:
    grade = 'A+'
elif marks >= 80:
    grade = 'A'
elif marks >= 70:
    grade = 'B'
elif marks >= 60:
    grade = 'C'
elif marks >= 50:
    grade = 'D'
else:
    grade = 'F'

print(f"Marks: {marks}, Grade: {grade}")
```

##### 18. Find the type of triangle based on side lengths

```python
# Given sides
a = 5
b = 5
c = 8

# Check the type of triangle
if a == b == c:
    triangle_type = "Equilateral triangle"
elif a == b or b == c or a == c:
    triangle_type = "Isosceles triangle"
else:
    triangle_type = "Scalene triangle"

print(f"Sides: {a}, {b}, {c} → {triangle_type}")
```

##### 19. Create a simple calculator using conditions

```python
# Given numbers
num1 = 10
num2 = 5

# Operation
operation = '+'  # Change to '-', '*', '/', '%'

# Perform calculation
if operation == '+':
    result = num1 + num2
elif operation == '-':
    result = num1 - num2
elif operation == '*':
    result = num1 * num2
elif operation == '/':
    if num2 != 0:
        result = num1 / num2
    else:
        result = "Error! Division by zero."
elif operation == '%':
    if num2 != 0:
        result = num1 % num2
    else:
        result = "Error! Division by zero."
else:
    result = "Invalid operation"

print(f"{num1} {operation} {num2} = {result}")
```

##### 20. Check whether a number is a single-digit, double-digit, or multi-digit
